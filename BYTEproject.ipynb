```python
import numpy as np
import pandas as pd
```

```python
# Loading the data provided

file_path = 'C:/Users/aryan/Downloads/Task_1.csv'


df = pd.read_csv(file_path)


print(df.head())
```

```python
import re
#Here I am removing all the speical symbols , spaces etc from the  data.
## So as to say the pre-processing of the data

def preprocess_text(text):
    text = text.lower()  
    text = re.sub(r'[^a-z\s]', '', text)  
    return text


df['text'] = df['text'].apply(preprocess_text)


print(df.head())
```

```python
## Trying to see what unique elements the labels have for example if it has spaces then that could be a problem 
print(df['labels'].unique())
```

```python
## making all the labels in lower case and zero space 
df['labels'] = df['labels'].str.strip().str.lower()
df['labels'] = df['labels'].map({'spam': 1, 'ham': 0})
```

```python
print(df.head())
```

```python
## SPlitting the data in training and testing data.
## Here I am using 80% data as training and rest as testing
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df['text'], df['labels'], test_size=0.2, stratify=df['labels'], random_state=42)
```

```python
## Vectorising the text messages into numerical form using TF-ID
## IT is important since machine understands only numnerical values
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer()
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)
```

```python
## We are going to use MultiNomial Naive Bayes Model 
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(X_train_tfidf, y_train)
```

```python
## Confusion Matrix test 
from sklearn.metrics import classification_report, accuracy_score
y_pred = model.predict(X_test_tfidf)
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```
